# Data Visualization

```python
sns.pairplot(task3)
```

```python
sns.pairplot(task3, hue='Region')
```

```python
fig = plt.figure(figsize = (40,40))
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

ax = fig.gca()
task3.hist(ax = ax)
```

```python
plt.figure(figsize = (15,15))
sns.countplot(x = task3['State'])
plt.xticks(rotation = 90)
plt.title("State")
plt.show()
```

```python
plt.figure(figsize = (10,10))
task3['Region'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

```python
sns.set(style = "darkgrid")
plt.figure(2, figsize = (20,15))
sns.barplot (x = 'Sub-Category', y = 'Discount', data=task3, palette = 'Spectral')
plt.suptitle('Sub-Category wise discount', fontsize = 16)
plt.show()
```

```python
task3.groupby('Region')[['Sales', 'Quantity','Profit']].sum().plot.bar(color=['blue','orange','green'],figsize=(8,5))
plt.ylabel('Profit gain and sales')
plt.show()
```

```python
ps = task3.groupby('Sub-Category')[['Sales','Quantity','Profit']].sum().sort_values(by = 'Sales', ascending = False)
ps[:].plot.bar(color = ['blue','green','orange'],figsize=(15,25))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & sales')
plt.show()
```

# Conclusion

1. Phones and Chairs Sub-Category has the large number of sales but has second and fourth highest profit margin
2. For the Tables, Bookcases and Supplies has the fourth, ninth and thirteenth number of sales in it but has negative Profit
3. Fasteners has the lowest sales but has the minimum profit among all others except Tables, Bookcases, Supplies.
4. Envelopes, Labels and Fasteners contributes the least towards sales and profit.

##### As a Business Manager, he/she should give more attention to the sub-categories areas of Tables, bookcases, and the supplies by which the negative profit would be reduced. Also on to the Fasteners subcategory to increase the profit  margin at that part.
